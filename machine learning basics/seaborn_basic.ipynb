```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
tip_data=sns.load_dataset('tips')
```

```python
tip_data.columns
```

```python
tip_data.day.unique()
```

```python
tip_data.sex.unique()
```

```python
sns.set_style('darkgrid')
sns.violinplot(x='tip',data=tip_data,color='b')
plt.title("total bill")
plt.show()
```

```python
#categorical data
iris=sns.load_dataset("iris")
sns.swarmplot(x="species",y="petal_length",data=iris)
```

```python
iris.species.unique()
```

```python
bx=sns.boxplot(x="day",y="total_bill",data=tip_data)#boxplot representation of data is used to check the quatile1, quatile2 as well as it shows outliers in data
bx.set_title("boxplot")
plt.show()
```

```python
bx=sns.boxplot(x="species",y="petal_length",data=iris)#boxplot representation of data is used to check the quatile1, quatile2 as well as it shows outliers in data
bx.set_title("boxplot")
plt.show()
```
